```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from prophet import Prophet
```

```python
df = pd.read_csv('actual.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['day_of_week'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter
```

```python
df.tail()
```

```python
df.sort_values(by='date')
```

```python
split_date = pd.to_datetime("2023-01-07")

df_training = df.loc[df['date'] < split_date]
df_test = df.loc[df['date'] >= split_date]
```

```python
df_test.tail()
```

```python
# Prepare data for Prophet
prophet_data = df_training.rename(columns={'date': 'ds', 'count': 'y'})
```

```python
# Initialize and fit the Prophet model for each user
sources = df_training['source'].unique()
```

```python
pSource = []
pNextDate = []
pNextCount = []
for source in sources:
    source_df = prophet_data[df_training['source'] == source][['ds', 'y']]
    model = Prophet()
    if(len(source_df)>5):
        model.fit(source_df)
        # Create a dataframe for today's date
        future = model.make_future_dataframe(periods=1, freq='D')
        # Predict uploads for today
        forecast = model.predict(future)
        # Check if the forecasted value exceeds a threshold (you can set this threshold)
        next_upload_date = forecast.loc[len(source_df), 'ds']
        next_upload_count_prediction = forecast.loc[len(source_df), 'yhat']
        pSource.append(source)
        pNextCount.append(next_upload_count_prediction)
        pNextDate.append(next_upload_date)
predictions = pd.DataFrame({'source': pSource,
     'next_count': pNextCount,
     'next_date': pNextDate
    })
```

```python
# List of users predicted to upload today
test1 = predictions[predictions['next_date'] == '2023-01-07']
test1 = test1.rename(columns={'source': 'source', 'next_count': 'count', 'next_date': 'date'})
```

```python
actuals = df_test[df['date']=='2023-01-07']['source']
preds = test1['source']
```

```python
print("Missing values in predictions list:", (set(actuals).difference(preds)))
print("Additional values prediction list:", (set(preds).difference(actuals)))
```
