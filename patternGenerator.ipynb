```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
import numpy as np
import random
from datetime import datetime, timedelta

sourceSystem = ['CPXD','TPXD','MKXD', 'KKXD', 'NFXD', 'TLXD', 'MCXD', 'GLXD', 'NCXD', 'KNXD', 'MTXD']

# Define the date range
start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 1, 10)

# Generate the date range with only weekdays
date_range = []
current_date = start_date

while current_date <= end_date:
    if current_date.weekday() < 5:  # Check if the day is a weekday (0 to 4 for Monday to Friday)
        date_range.append(current_date)
    current_date += timedelta(days=1)
df = pd.DataFrame()
# Generate random IDs
for x in sourceSystem:
    random_ids = [x] * len(date_range)
    df = df.append(pd.DataFrame({'source': random_ids, 'Date': date_range}))

# Display the DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load your time series data
data = pd.read_csv('sample_data.csv')
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['source']
```

```python
# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['source'], label='Original Data')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plot_acf(data['source'], lags=40)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(data['value'], lags=40)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.grid(True)
plt.show()
```
